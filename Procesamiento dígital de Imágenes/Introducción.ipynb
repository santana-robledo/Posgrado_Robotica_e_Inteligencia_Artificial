```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
from google.colab.patches import cv2_imshow
I = cv2.imread("/content/Ingesaurio.jpeg")
cv2_imshow(I)
```

```python
print("Resolucion original",I.shape)

new_width=100;
new_height=100;

resized=cv2.resize(I,(new_width,new_height))
print("Resolucion redimensionada",resized.shape)

cv2_imshow(I)
print("")
cv2_imshow(resized)
```

```python
cm = cv2.VideoCapture(0)

# Check if camera opened successfully
if not cm.isOpened():
    print("Error: Could not open video stream or file")
else:
    while True:
        ret, I = cm.read()

        # Check if frame is read successfully
        if not ret or I is None:
            print("Error: Could not read frame from camera. Exiting.")
            break

        cv2_imshow(I)

        if cv2.waitKey(10) == ord("q"):
            break

    cm.release()
    cv2.destroyAllWindows()
```
