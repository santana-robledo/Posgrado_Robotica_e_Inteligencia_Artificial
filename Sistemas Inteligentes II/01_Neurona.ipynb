```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
class Neurona:
    def __init__(self,n_inputs):
        self.w= -1 + 2*np.random.rand(n_inputs)
        self.b= -1 + 2*np.random.rand()

    def predict (self,X):
        _,p=X.shape

        y_est=np.zeros(p)

        for i in range(p):
            y_est[i]=np.dot(self.w,X[:,i])+self.b
            if y_est[i]>0:
                y_est[i]=1
            else:
                y_est[i]=0
        return y_est
```

```python
class Perceptron:

  def __init__(self,n_inputs,learning_rate):
    self.w=-1+2*np.random.rand(n_inputs)
    self.b=-1+2*np.random.rand()
    self.lr=learning_rate

  def predict(self,X):
    _,p=X.shape
    y_est=np.zeros(p)
    for i in range(p):
      y_est[i]=np.dot(self.w,X[:,i])+self.b
      if y_est[i]>0:
          y_est[i]=1
      else:
          y_est[i]=0
    return y_est

  def fit(self,X,Y,epochs=50):
      _,p=X.shape
      for _ in range(epochs):
        for i in range(p):
          y_est=self.predict(X[:,i].reshape(-1,1))
          self.w+=self.lr *(Y[i]-y_est)*X[:,i]
          self.b += self.lr *(Y[i]-y_est)
```

```python
def draw_2d_percep(model):
  w1, w2, b = model.w[0], model.w[1], model.b
  plt.plot([-2, 2],[(1/w2)*(-w1*(-2)-b),(1/w2)*(-w1*2-b)],'--k')
```

```python
# Instanciar el modelo
model = Perceptron(2, 0.1)

# Datos
X = np.array([[0, 0, 1, 1],
              [0, 1, 0, 1]])
Y = np.array( [0, 1, 1, 1])

# Entrenar
model.fit(X,Y)

# Predicción
model.predict(X)

# Primero dibujemos los puntos
_, p = X.shape
for i in range(p):
  if Y[i] == 0:
    plt.plot(X[0,i],X[1,i], 'or')
  else:
    plt.plot(X[0,i],X[1,i], 'ob')

plt.title('Compuerta OR')
plt.grid('on')
plt.xlim([-2,2])
plt.ylim([-2,2])
plt.xlabel(r'x1')
plt.ylabel(r'x2')

draw_2d_percep(model)
```
